### <a name="mark_01"></a> Generalidades.
### [Index](#index)


Originalmente las bases de datos se utilizaban de manera texto, es decir, que se guardaban los datos de manera plana, en archivos de texto ‘.txt’. El problema era que al momento de sacarlos, leerlos y hacer cosas interesantes con los datos se complicaba.

En ese contexto nacen las bases de datos relacionales fueron un esfuerzo, incluido matemático, para hacer la conjunción de datos y mantenerlos estables, de forma que no solo pudiéramos guardar, sino también extraer y hacer cosas ingeniosas con los datos, como cálculos estadísticos. Ejemplos de estas bases de datos serían: SQLServer, Oracle, PostgreSQL, MariaDB, MySQL.

Las bases de datos relacionales históricamente son una navaja Suiza, son la herramienta que arregla en general todos los casos de datos y te permite hacer sistemas muy robustos con una sola herramienta.

Por otro lado empezó un problema histórico a inicio de los años 2000, en el que teníamos web 2.0, bases de datos gigantes y redes sociales. Fue un gran torrente de datos incrementándose día a día. Para esto se crearon ciencias para manejo de grandes cantidades de datos, como BIG DATA. En respuesta a esto se crearon nuevas bases de datos que fueron específicas para manejar ciertos casos de datos.

Las Bases de Datos No Relacionales son un tipo diferente de ddbb. Algunas basadas en documentos, otras basadas en grafos, otras son columnares, algunas se las guarda en memoria.

Las bases de datos no relacionales, son herramientas específicas para ciertos tipos de trabajo como: mantener el estado de la aplicación (base de datos basado en documentos), si se quiere hacer analítica de datos se requiere un datawarehouse (algo basado en grafos), si se quiere tener algo con relaciones muy complejas entre entidades se recomienda las bases de datos basadas en grafos. Dependiendo el caso de uso se utiliza una herramienta diferente.

Las bases de datos relacionales también evolucionaron con el paso del tiempo, se han acoplado nuevas funcionalidades, como el implemento de funciones, nuevos esquemas de trabajo y varias funcionalidades nuevas para mantenerse vivas en el mundo del manejo de datos.

create or replace function count_total_movies()
--debe retornar un valor.
returns int

as $$
begin

En la sentencia PL/pgSQL que proporcionas, "returns" y "return" cumplen funciones distintas:

1. "returns int": Esta línea se encuentra en la declaración de la función y especifica el tipo de dato que la función va a retornar. En este caso, "int" indica que la función retornará un valor entero (integer).

2. `return count(*) from peliculas`: Esta línea está dentro del bloque de la función, entre el "begin" y "end". Indica que la función va a retornar el resultado de la consulta que cuenta el número de filas en la tabla "peliculas". La función "count(*)" cuenta todas las filas de la tabla y el resultado será un valor entero.

En resumen, "returns int" se utiliza para declarar el tipo de dato que la función va a retornar, mientras que "return" se utiliza dentro del bloque de la función para especificar el valor que será retornado cuando la función sea llamada. En este caso, la función va a retornar el resultado de la consulta que cuenta las filas de la tabla "peliculas", lo que dará como resultado un valor entero.

Al ejecutarlo esta función se ubicará en **Servers --> PostgresSQL 14 --> Databases --> platzimovies --> Schemas --> Functions**, listo para ser invocado.

## Diferencias entre "procedures" y "functions":

En PostgreSQL, tanto los procedimientos almacenados (`PROCEDURE`) como las funciones (`FUNCTION`) son objetos de base de datos que contienen código SQL y PL/pgSQL (PL/pgSQL es un lenguaje procedural específico de PostgreSQL). Ambos se utilizan para encapsular lógica empresarial y permitir su reutilización.

Aquí hay algunas diferencias clave entre procedimientos almacenados y funciones en PostgreSQL:

1. **Valor de Retorno:**
   - **Funciones:** Deben devolver un valor. Pueden devolver un solo valor o una tabla (conjunto de filas).
   - **Procedimientos Almacenados:** Pueden o no devolver un valor. Pueden tener parámetros de salida, pero no están diseñados específicamente para devolver valores como las funciones.

   **Ejemplo:**
   ```sql
   -- Función que devuelve un solo valor
   CREATE OR REPLACE FUNCTION obtener_suma(a INT, b INT) RETURNS INT AS $$
   BEGIN
      RETURN a + b;
   END;
   $$ LANGUAGE plpgsql;

   -- Procedimiento almacenado sin valor de retorno explícito
   CREATE OR REPLACE PROCEDURE imprimir_mensaje(mensaje TEXT) AS $$
   BEGIN
      RAISE NOTICE '%', mensaje;
   END;
   $$ LANGUAGE plpgsql;
   ```

2. **Transacciones Implícitas:**
   - **Funciones:** Pueden ejecutar transacciones implícitas y pueden incluir operaciones de lectura y escritura en la base de datos.
   - **Procedimientos Almacenados:** No inician automáticamente transacciones. Deben ejecutarse dentro del contexto de una transacción más grande.

3. **Uso en Consultas:**
   - **Funciones:** Pueden ser utilizadas en consultas SQL como parte de la expresión. Pueden ser invocadas en la cláusula `SELECT`, en condiciones `WHERE`, etc.
   - **Procedimientos Almacenados:** Se ejecutan usando la sintaxis específica de procedimientos almacenados y generalmente no se usan directamente en consultas SQL.

   **Ejemplo:**
   ```sql
   -- Uso de una función en una consulta
   SELECT obtener_suma(5, 3);

   -- Invocación de un procedimiento almacenado
   CALL imprimir_mensaje('Hola, mundo');
   ```

4. **Ejecución de Acciones:**
   - **Funciones:** Pueden realizar acciones y devolver resultados.
   - **Procedimientos Almacenados:** Están diseñados principalmente para ejecutar acciones. Pueden no devolver un resultado explícito.

5. **Ventajas de Performance:**
   - En términos de rendimiento, las funciones son a menudo más eficientes, ya que se pueden optimizar más fácilmente. PostgreSQL puede hacer un mejor trabajo al planificar consultas que involucran funciones.
  
   **Ejemplo de Optimización:**
   ```sql
   -- Uso de una función en una consulta
   SELECT * FROM mi_tabla WHERE columna = obtener_valor_optimizado();
   ```

   En este ejemplo, PostgreSQL puede optimizar la ejecución de la función y planificar la consulta de manera más efectiva.

En resumen, las funciones son más adecuadas cuando necesitas devolver un resultado calculado, mientras que los procedimientos almacenados son útiles para realizar acciones o tareas procedurales sin necesidad de devolver un valor específico. La elección entre ambos dependerá de los requisitos específicos de tu aplicación.

### <a name="mark_10"></a>PLPGSQL: Triggers.
### [Index](#index)

******************** El trigger Primer Ejemplo **************************

El primer paso es crear una función que se ejecute cada vez que una acción especifica sea realizada.

A continuación, se presenta un ejemplo de creación de un trigger que se ejecuta después de cada inserción en una tabla "empleados" y actualiza automáticamente el campo "fecha_actualizacion" con la fecha y hora actual:

Acá creamos la función "actualizar_fecha()" que retorna un "TRIGGER", ya que esta función se dispará luego de que se inserte un nuevo empleado a la tabla "empleados"
```sql
CREATE OR REPLACE FUNCTION actualizar_fecha()
RETURNS TRIGGER AS
$$
BEGIN
    NEW.fecha_actualizacion := CURRENT_TIMESTAMP;
    RETURN NEW;
END;
$$
LANGUAGE plpgsql;

CREATE TRIGGER trigger_actualizar_fecha
AFTER INSERT ON empleados
FOR EACH ROW
EXECUTE FUNCTION actualizar_fecha();
```

En este ejemplo, se crea una función llamada "actualizar_fecha()" que se utilizará como acción del trigger. La función recibe el argumento `NEW`, que representa la fila que se está insertando en la tabla. Luego, se actualiza el campo "fecha_actualizacion" de la fila con el valor de la fecha y hora actual mediante `NEW.fecha_actualizacion := CURRENT_TIMESTAMP;`.


La cláusula "CREATE TRIGGER" en SQL se utiliza para crear un disparador o trigger en una base de datos. Un trigger es un objeto de base de datos que se ejecuta automáticamente cuando ocurre un evento específico en una tabla o vista. Los eventos pueden ser operaciones de inserción (INSERT), actualización (UPDATE) o eliminación (DELETE) de registros en la tabla.

La sintaxis general para crear un trigger es la siguiente:

```sql
CREATE TRIGGER nombre_trigger
{BEFORE | AFTER | INSTEAD OF} {INSERT | UPDATE | DELETE} ON nombre_tabla
[FOR EACH ROW]
WHEN (condicion)
EXECUTE FUNCTION nombre_funcion;
```

Donde:
- `nombre_trigger`: Es el nombre del trigger que se está creando.
- `{BEFORE | AFTER | INSTEAD OF}`: Especifica cuándo se ejecutará el trigger. "BEFORE" indica que el trigger se ejecutará antes de que se realice la acción, "AFTER" indica que se ejecutará después, e "INSTEAD OF" se utiliza en vistas.
- `{INSERT | UPDATE | DELETE}`: Indica sobre qué tipo de evento se ejecutará el trigger.
- `nombre_tabla`: Es el nombre de la tabla a la que se asociará el trigger.
- `[FOR EACH ROW]`: Es opcional y se utiliza para especificar que el trigger se ejecutará por cada fila afectada por el evento.
- `WHEN (condicion)`: Es opcional y se utiliza para especificar una condición que debe cumplirse para que se ejecute el trigger.
- `EXECUTE FUNCTION nombre_funcion`: Especifica la función o procedimiento que se ejecutará cuando se active el trigger.

El trigger se asocia a la tabla "empleados" con el evento "AFTER INSERT", lo que significa que se ejecutará después de cada inserción en la tabla. Cada vez que se inserte una nueva fila en la tabla "empleados", el trigger se activará y actualizará automáticamente el campo "fecha_actualizacion".

******************** El trigger Segundo Ejemplo **************************

Funcion a ser disparada:

```sql
create or replace function duplicate_records()
returns trigger
language plpgsql
as $$
begin
    insert into table_02(columna_01_b, columna_02_b)
    values(new.column_01_a, new.column_02_a );
    return new;
end
$$
```
**Importante!!!**: En la función "duplicate_records()" en ningún momento aparece "table_01", ésta aparece en la creación del Trigger donde se indica que tabla disparará la acción.

Al ejecutarlo esta función, esta se ubicará en **Servers --> PostgresSQL 14 --> Databases --> platzimovies --> Schemas --> Trigger Functions**, listo para ser invocado.

El código proporcionado es un ejemplo de una función trigger en PL/pgSQL que se utiliza para duplicar registros de una tabla "table_01" en otra tabla "table_02" cuando se realiza una operación de inserción en "table_01". Veamos paso a paso qué hace esta función trigger:

1. `create or replace function duplicate_records()`: Aquí se define una nueva función llamada "duplicate_records" que será utilizada como trigger.

2. `returns trigger`: La función se define con el tipo de retorno "trigger". Esto indica que la función se utilizará como un trigger y devolverá un valor de tipo "trigger".

3. `language plpgsql`: Se especifica que la función está escrita en PL/pgSQL, que es un lenguaje procedural específico para PostgreSQL.

4. `as $$`: Se inicia el bloque de código de la función trigger.

5. `begin`: Se inicia el bloque de código de la función.

6. `insert into table_02(columna_01_b, columna_02_b)`: Se realiza una instrucción de inserción en la tabla "table_02".

7. `values(new.column_01_a, column_02_a)`: Se especifican los valores que se insertarán en la tabla "table_02". "new" es una variable especial que contiene la fila que se está insertando en la tabla "table_01". Aquí se utiliza "new.column_01_a" para obtener el valor de la columna "columna_01_a" de la fila que se está insertando en "table_01", y "column_02_a" para obtener el valor de la columna "columna_02_a" de la misma fila.

8. `return new;`: Se devuelve la fila que se ha insertado en "table_01". Esto es necesario porque es un trigger "AFTER INSERT", lo que significa que el trigger se ejecuta después de la inserción en "table_01". Devolver "new" asegura que el proceso de inserción en "table_01" continúe sin problemas.

9. `end`: Se finaliza el bloque de código de la función.

10. `$$`: Se cierra el bloque de código de la función trigger.

En resumen, esta función trigger se ejecutará después de cada inserción en la tabla "table_01". Cuando se inserta una nueva fila en "table_01", el trigger toma los valores de las columnas "columna_01_a" y "columna_02_a" de la nueva fila y los inserta en "table_02". Luego, devuelve la nueva fila para que la inserción en "table_01" se complete sin problemas. De esta manera, se logra duplicar registros de una tabla a otra cuando se insertan nuevos registros.

* Creación del trigger:

```sql
create trigger table_01_changes
	before insert
	on table_01
	for each row --por cada registro nuevo
	execute function duplicate_records();
```
Al ejecutarlo esta trigger, esta se ubicará en **Servers --> PostgresSQL 14 --> Databases --> platzimovies --> Schemas --> Tables --> table_01 --> Triggers**, listo para ser invocado.

### <a name="mark_11"></a>PLPGSQL: Función Con Datos Estadísticos.
### [Index](#index)

```sql
--función estadística aplicada a la ciencia de datos:
create or replace function movies_stats()
returns void
language plpgsql
as $$
declare
	total_rated_r real := 0.0;
	total_larger_than_100 real := 0.0;
	total_published_2006 real := 0.0;
	average_duration real := 0.0;
	average_rental_price real := 0.0;
begin
    total_rated_r := count(*) from peliculas where clasificacion='R';
	total_larger_than_100 := count(*) from peliculas where duracion>100;
	total_published_2006 := count(*) from peliculas where anio_publicacion=2006;
	average_duration := avg(duracion) from peliculas;
	average_rental_price := avg(precio_renta) from peliculas;
	
	truncate table peliculas_estadisticas;
	insert into peliculas_estadisticas(tipo_estadistica, total)
	values
		('Total Clase R', total_rated_r),
		('Total Mayores a 100 min', total_larger_than_100),
		('Total Publicadas en 2006', total_published_2006),
		('Promedio duración en min', average_duration),
		('Promedio precio de renta', average_rental_price);
end
$$;

--llamamos a la nueva función
select movies_stats()

--checkeamos los cambios
select*from peliculas_estadisticas
```
* Nota_01: 
    - El operador **:=** se utiliza para inicializar una variable en este caso con un flotante **0.0**

* Nota_02:

    - La sentencia `TRUNCATE TABLE` se utiliza en SQL para eliminar todos los registros de una tabla de forma rápida y eficiente. A diferencia de la sentencia `DELETE`, que borra los registros uno por uno y puede ser lenta para grandes cantidades de datos, `TRUNCATE TABLE` es mucho más rápida porque elimina todos los datos de la tabla en una sola operación.

    - Al utilizar `TRUNCATE TABLE`, los datos se eliminan por completo y no se pueden recuperar. Esta sentencia es ideal cuando se desea vaciar completamente una tabla y se tiene la certeza de que no se necesitará recuperar los datos eliminados. También es útil para reiniciar tablas que almacenan datos temporales o que necesitan ser restablecidas periódicamente.

    - Es importante tener en cuenta que `TRUNCATE TABLE` es una operación DDL (Data Definition Language) en lugar de una operación DML (Data Manipulation Language) como `DELETE`. Esto significa que no se registra cada eliminación en el log de transacciones y no activa los disparadores (triggers) de borrado, lo que contribuye a su mayor velocidad en comparación con `DELETE`.

La sintaxis de `TRUNCATE TABLE` es simple:

```
TRUNCATE TABLE nombre_tabla;
```

Donde `nombre_tabla` es el nombre de la tabla que se desea truncar. Es importante tener los permisos adecuados para ejecutar esta operación, ya que eliminará todos los datos de la tabla sin posibilidad de recuperación. Por lo tanto, asegúrate de tener una copia de seguridad de los datos antes de utilizar `TRUNCATE TABLE`.

* Profundizando:

1. **Operación DDL (Data Definition Language):** En SQL, las operaciones DDL se utilizan para definir o modificar la estructura de la base de datos, como crear tablas, eliminar tablas, modificar columnas, entre otras. `TRUNCATE TABLE` es una operación DDL porque afecta la estructura de la tabla al eliminar todos los datos contenidos en ella, pero no modifica la definición de la tabla en sí.

2. **Operación DML (Data Manipulation Language):** Por otro lado, las operaciones DML se utilizan para manipular los datos almacenados en la base de datos, como insertar, actualizar o eliminar registros. `DELETE` es una operación DML porque se encarga de eliminar registros específicos de la tabla, lo que afecta directamente los datos almacenados en ella.

3. **Registro en el log de transacciones:** Cuando se realiza una operación DML, como `DELETE`, se registra cada eliminación en el log de transacciones de la base de datos. Esto es importante para asegurar que los cambios se puedan deshacer en caso de una falla del sistema o para cumplir con ciertas políticas de recuperación de datos. Sin embargo, al utilizar `TRUNCATE TABLE`, no se registra cada eliminación individual, lo que significa que la operación es más rápida, ya que no hay necesidad de escribir cada eliminación en el log.

4. **Disparadores (triggers) de borrado:** Los disparadores son procedimientos almacenados que se ejecutan automáticamente cuando ocurre un evento específico en una tabla, como una inserción, actualización o eliminación. Los disparadores de borrado se ejecutan cuando se realiza la operación `DELETE`, pero no se activan cuando se utiliza `TRUNCATE TABLE`. Esto puede ser útil en situaciones donde se desea realizar acciones adicionales antes o después de borrar datos de una tabla.

En resumen, `TRUNCATE TABLE` es más rápida que `DELETE` porque no registra cada eliminación en el log de transacciones y no activa los disparadores de borrado. Sin embargo, es importante tener en cuenta que `TRUNCATE TABLE` borra todos los datos de la tabla de una sola vez y no es reversible, por lo que debe usarse con precaución y asegurarse de que no se necesitará recuperar los datos eliminados.

### <a name="mark_16"></a>Common Table Expressions.
### [Index](#index)

[Documentación](https://www.postgresql.org/docs/12/queries-with.html)

* las CTEs comparten ciertas características con las funciones:

1. **Devuelve un resultado:** Al igual que una función, una CTE calcula y devuelve un conjunto de resultados basados en la lógica definida en la CTE. Estos resultados pueden ser utilizados en una consulta principal de la misma manera que se usaría una tabla real.

2. **Reutilización:** Al igual que una función, una CTE puede ser referenciada varias veces en la misma consulta. Esto permite que los resultados de la CTE sean utilizados en diferentes partes de la consulta principal sin necesidad de repetir la misma lógica de cálculo.

La principal diferencia entre una CTE y una función es que una CTE es una parte integral de la consulta SQL en sí misma, mientras que una función es un objeto separado que debe ser invocado y llamado explícitamente. Las CTEs son especialmente útiles cuando deseas estructurar y organizar tu consulta de manera más legible y mantenible, al tiempo que evitas la repetición de lógica.

En resumen, las CTEs son una forma conveniente de definir y utilizar subconjuntos de datos en una consulta SQL, mejorando la claridad y reutilización de la lógica de cálculo.

A veces, cuando estás programando, te puede pasar que tengas que hacer un bucle en Python, PHP, JS o el lenguajes que estés usando (a partir de ahora diré Python para simplificar) y que dentro del bucle tengas que ejecutar una sentencia SQL. Realizar cálculos en Python y volver a insertar datos. En estos casos, en cada vuelta del loop Python tendrá que llamar internamente al conector que le permite leer la base de datos, luego realizará los cálculos y conectará otra vez para con la DB para insertar el valor final.

Bueno, en estos casos puedes hacer uso de las Common Table Expressions para hacer tu loop directamente en PostgreSQL, realizar tus calculos y volver a insertar sin tener que llamar al conector en cada itineración. Ahorrarás tiempos, recursos, dinero, la aplicación será más rápida.

```sql
with recursive tabla_recursiva(num_column) as (
	values(1)
	union all
	select num_column+1 from tabla_recursiva where num_column < 100
) 
select sum(num_column) from tabla_recursiva;
```
* Nota:
**La "num_column" dentro de la definición de tabla_recursiva(num_column) es una columna que representa los valores de la secuencia numérica generada en la CTE recursiva. La CTE se define como una secuencia de valores numéricos comenzando desde 1 y aumentando en 1 en cada iteración, hasta que el valor de "n" alcanza 100.**

En cada iteración de la recursión, la consulta select num_column+1 from tabla_recursiva where num_column < 100 agrega una nueva fila a la CTE tabla_recursiva donde "num_column" se incrementa en 1 en comparación con el valor de la fila anterior. Esto crea una secuencia de números ascendentes que comienza en 1 y continúa hasta llegar a 100. Estos valores de "num_column" se utilizan posteriormente en la última consulta select sum(num_column) from tabla_recursiva; para calcular la suma total de los números en la secuencia generada.

Su equivalente en Python:
```python
a = 0
for i in range(101):
    a = i + a
```
Explicación paso a paso:

Consulta recursiva para generar una secuencia de números y luego calcula la suma de esos números. Aquí tienes una explicación paso a paso de lo que está sucediendo:

1. Se inicia con una cláusula `WITH RECURSIVE` que define la consulta recursiva y la nombra como "tabla_recursiva". Dentro de esta cláusula, comenzamos con un valor base utilizando `VALUES(1)`, que es el primer número de la secuencia.

2. Luego, utilizamos `UNION ALL` para combinar el valor base con el resultado de la siguiente parte de la consulta.

3. La siguiente parte de la consulta es una subconsulta que selecciona "num_column + 1" de "tabla_recursiva" donde "num_column" sea menor que 100. Esto agrega un nuevo valor a la secuencia incrementando el valor anterior en uno.

4. Como estamos usando una cláusula `UNION ALL`, la consulta recursiva se ejecuta repetidamente y agrega nuevos valores a la secuencia hasta que el valor "num_column" llegue a 100.

5. Después de definir la consulta recursiva, fuera de la cláusula `WITH RECURSIVE`, realizamos una consulta que suma todos los valores de la secuencia generada en "tabla_recursiva".

En resumen, este código SQL utiliza la recursión para generar una secuencia de números del 1 al 100 y luego calcula la suma de estos números. Cada iteración de la recursión agrega un nuevo valor a la secuencia, incrementando el valor anterior en uno. Finalmente, se calcula y devuelve la suma de todos los valores en la secuencia generada.

### <a name="mark_17"></a>Particiones.
### [Index](#index)

Particionado - 

Dividir una tabla en segmentos lógicos. 
Es una practica común de los manejadores, pero no todos ofrecen la opción para los usuarios de administrar estas particiones. 
Resulta útil para optimizar las consultas de datos.

Observaciones:

No todas las tablas deben de ser particionadas, vale la pena hacerlo unicamente cuando hay muchos registros.

Permite optimizar algunas consultas al no tener que buscar dentro de toda una tabla sino unicamente en un segmento especifico.

El particionado altera la consistencia de la tablas.

No existen los indices (llaves primarias) en las particiones, o mejor dicho, estos indices cambian basándose en la partición. e.g Si particionas una tabla por fechas, al buscar un dato especifico el primer criterio de búsqueda será la fecha.

El particionado en PostgreSQL es una técnica que divide una gran tabla en partes más pequeñas llamadas particiones. Cada partición tiene sus propias características y puede almacenarse en un lugar diferente. Aquí hay un ejemplo básico de cómo puedes crear una tabla particionada en PostgreSQL:

Supongamos que tienes una tabla llamada `ventas` que contiene información sobre las ventas, y deseas particionarla por fecha. Aquí hay un ejemplo de cómo podrías hacerlo:

```sql
-- Crear una tabla maestra para definir la estructura de particionado
CREATE TABLE ventas (
    id serial PRIMARY KEY,
    fecha_venta DATE,
    monto NUMERIC
);

-- Crear las particiones
CREATE TABLE ventas_2019 PARTITION OF ventas
    FOR VALUES FROM ('2019-01-01') TO ('2020-01-01');

CREATE TABLE ventas_2020 PARTITION OF ventas
    FOR VALUES FROM ('2020-01-01') TO ('2021-01-01');

CREATE TABLE ventas_2021 PARTITION OF ventas
    FOR VALUES FROM ('2021-01-01') TO ('2022-01-01');

-- Crear un índice en la columna de partición
CREATE INDEX idx_fecha_venta ON ventas (fecha_venta);

-- Insertar datos en las particiones
INSERT INTO ventas_2019 (fecha_venta, monto) VALUES ('2019-05-15', 1000);
INSERT INTO ventas_2020 (fecha_venta, monto) VALUES ('2020-03-20', 1500);
INSERT INTO ventas_2021 (fecha_venta, monto) VALUES ('2021-08-10', 2000);

-- Consulta de ejemplo
SELECT * FROM ventas WHERE fecha_venta >= '2020-01-01' AND fecha_venta < '2021-01-01';
```

En este ejemplo, creamos una tabla principal llamada `ventas` y luego creamos tres tablas de particiones (`ventas_2019`, `ventas_2020`, `ventas_2021`). Estas particiones contienen datos específicos según el rango de fechas definido en las cláusulas `FOR VALUES`.

Es importante destacar que el particionado en PostgreSQL se basa en el rango de valores de la columna de partición. También se puede utilizar particionado por lista o hash, dependiendo de los requisitos específicos del caso de uso.

Las funciones de ventana (window functions) en PostgreSQL son funciones analíticas que operan en un conjunto de filas relacionadas con la fila actual, conocido como "ventana". Estas funciones son extremadamente útiles para realizar cálculos comparativos y resúmenes en un conjunto de datos. A continuación, te proporcionaré una explicación paso a paso junto con ejemplos.

### Paso 1: Crear una tabla de ejemplo

```sql
CREATE TABLE ventas (
    id serial PRIMARY KEY,
    fecha_venta DATE,
    monto NUMERIC
);

INSERT INTO ventas (fecha_venta, monto) VALUES 
    ('2022-01-01', 1000),
    ('2022-01-02', 1500),
    ('2022-01-03', 2000),
    ('2022-01-04', 1200),
    ('2022-01-05', 1800),
    ('2022-01-01', 1001),
    ('2022-01-02', 1501),
    ('2022-01-03', 2001),
    ('2022-01-04', 1201),
    ('2022-01-05', 1801);
```

### Paso 2: Uso básico de una función de ventana

Vamos a utilizar la función `ROW_NUMBER()` para asignar un número de fila a cada fila en función de la fecha de venta.

```sql
SELECT 
    id,
    fecha_ventas,
    monto,
    ROW_NUMBER() OVER (ORDER BY fecha_ventas) AS numero_fila
FROM ventas;
```

En este ejemplo, `ROW_NUMBER()` asignará un número secuencial a cada fila basado en la ordenación de la columna `fecha_ventas`.

![](img_01.png)

### Paso 3: Uso de más de una función de ventana

Vamos a usar tanto `ROW_NUMBER()` como `SUM()` para calcular la suma acumulativa de los montos.

```sql
SELECT 
    id,
    fecha_ventas,
    monto,
    ROW_NUMBER() OVER (ORDER BY fecha_ventas) AS numero_fila,
    SUM(monto) OVER (ORDER BY fecha_ventas) AS suma_acumulativa
FROM ventas;
```

![](img_02.png)

En este caso, `ROW_NUMBER()` sigue asignando números secuenciales, mientras que `SUM(monto) OVER (ORDER BY fecha_ventas)` calcula la suma acumulativa de los montos basándose en la ordenación de la columna `fecha_venta`.

### Paso 4: Especificar el marco de la ventana

La cláusula `ROWS` te permite especificar el rango exacto de filas para cada cálculo.

```sql
SELECT 
    id,
    fecha_ventas,
    monto,
    ROW_NUMBER() OVER (ORDER BY fecha_ventas) AS numero_fila,
    SUM(monto) OVER (ORDER BY fecha_ventas ROWS BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW) AS suma_acumulativa
FROM ventas;
```

![](img_03.png)

Aquí, `ROWS BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW` significa que la ventana incluirá todas las filas desde el principio hasta la fila actual.

Estos son solo ejemplos básicos. Las funciones de ventana ofrecen una gran flexibilidad y son especialmente útiles en situaciones más complejas, como cálculos de ranking, promedios móviles, entre otros.

Supongamos que queremos calcular el top 10 de las peliculas más vendidas en un Blockbuster

La siguiente query podría ayudar como ejemplo.

```sql
select 
	peliculas.pelicula_id,
	peliculas.titulo,
	count(*) as numero_rentas,
	--window function init
    row_number() over (
		order by count(*) desc
	) as ranking
    --window function fin
from rentas
	inner join inventarios on inventarios.inventario_id = rentas.inventario_id
	inner join peliculas on peliculas.pelicula_id = inventarios.pelicula_id
group by peliculas.pelicula_id
order by numero_rentas desc
limit 10
```

Consulta paso a paso:

1. **Seleccionar las columnas:** `pelicula_id` y `titulo` de la tabla `peliculas`.

2. **Calcular el número de rentas:** Usa la función de agregación `count(*)` para contar cuántas veces aparece cada película en la tabla `rentas`. El resultado de esta cuenta se nombra como `numero_rentas` en la salida.

3. **Calcular el ranking:** Usa la función de ventana `row_number()` para asignar un rango a cada fila, basado en el número de rentas (`count(*)`) en orden descendente.
